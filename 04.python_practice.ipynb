```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
browser = webdriver.Chrome()
```

```python
url = 'https://play.google.com/store/apps/details?id=com.iloen.melon&hl=ko-KR'
browser.get(url)
```

```python
browser.find_element(By.XPATH, '//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[2]/div[2]/div/div[1]/div[1]/c-wiz[4]/section/header/div/div[2]/button').click()
```

```python
container = browser.find_elements(By.CLASS_NAME, 'RHo1pe')

date_list = []
star_list = []
comment_list =[]
useful_list = []

for i in container:
    # 게시일 가져오기
    dates = i.find_element(By.CLASS_NAME, 'bp9Aid').text
    date_list.append(dates)

    # 별점
    stars = i.find_element(By.CLASS_NAME, 'iXRFPc').get_attribute('aria-label')[10]
    star_list.append(stars) 

    # 댓글 내용 가져오기
    comments = i.find_element(By.CLASS_NAME,'h3YV2d').text
    comment_list.append(comments)

    # 추천/유용
    try:
        useful = i.find_element(By.CLASS_NAME,'AJTPZc').text
        useful_list.append(useful)
    except:
        useful_list.append("사용자 0명이 이 리뷰가 유용하다고 평가함")

data = {
    "날짜" : date_list,
    "별점" : star_list,
    "댓글 내용" : comment_list,
    "유용/추천" : useful_list
}

import pandas as pd
df = pd.DataFrame(data)
df.to_csv("멜론 앱 댓글 크롤링.csv") 
df
```
