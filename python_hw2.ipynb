## 네이버 뉴스 크롤링
- 제목/내용/언론/링크
- 엑셀 저장

```python
# 네이버 뉴스 제목/내용/언론/링크
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
browser = webdriver.Chrome()
```

```python
url = 'https://news.naver.com/main/list.naver?mode=LS2D&mid=shm&sid1=101&sid2=259'
browser.get(url)
```

```python
frame = browser.find_element(By.CLASS_NAME, 'list_body')

containers = frame.find_elements(By.TAG_NAME, 'li')

title_list = []
content_list = []
company_list = []
link_list = []

for item in containers:
    # 제목 가져오기
    title = item.find_element(By.TAG_NAME, 'img').get_attribute('alt')
    title_list.append(title)

    # 내용 가져오기
    content = item.find_element(By.CLASS_NAME, 'lede').text
    content_list.append(content)

    # 언론사 가져오기
    company = item.find_element(By.CLASS_NAME, 'writing').text
    company_list.append(company)

    # 링크 가져오기
    link = item.find_element(By.TAG_NAME, 'a').get_attribute('href')
    link_list.append(link)
```

```python
NaverNews = {
    "제목" : title_list,
    "내용" : content_list,
    "언론사" : company_list,
    "링크" : link_list
}

df = pd.DataFrame(NaverNews)
df

df.to_csv("네이버 금융 뉴스 크롤링.csv")
```

## 예스24 베스트셀러 크롤링
- 책 제목, 저자, 출간일
- 엑셀 저장

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
browser = webdriver.Chrome()
```

```python
url = 'https://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=06'
browser.get(url)
```

```python
book_info = browser.find_elements(By.CLASS_NAME, 'goodsTxtInfo')

title_list = []
link_list = []
writer_list = []
company_list = []
publish_list = []

for item in book_info:
    # 제목 가져오기
    title = item.find_element(By.TAG_NAME, 'a').text
    title_list.append(title)

    # 링크 가져오기
    link = item.find_element(By.TAG_NAME, 'a').get_attribute('href')
    link_list.append(link)

    # 저자 가져오기
    info = item.find_element(By.CLASS_NAME, 'aupu').text
    info_list = info.split(" | ")

    writer = info_list[0]
    writer_list.append(writer)

    company = info_list[1]
    company_list.append(company)

    publish = info_list[2]
    publish_list.append(publish)

data = {
    "제목" : title_list,
    "저자" : writer_list,
    "출판사" : company_list,
    "출판일" : publish_list,
    "링크" : link_list
}

df = pd.DataFrame(data)
df.to_csv("yes24_bestseller.csv")
df
```
